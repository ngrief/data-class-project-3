```python
top10_country=data_df.loc[
    data_df['country'].isin([
        'India', 'Mexico', 'Russian Federation', 'China', 'Bangladesh', 'Pakistan', 'Ukraine', 'Philippines', 'Poland', 'United Kingdom'
    ])]
top10_country
```

```python
# Calculate the correlation between population density and net migration
correlation = top10_country['pop_density'].corr(top10_country['migration_perc'])

correlation
```

The correlation between population density and migration percentage is approximately -0.25, indicating a weak negative relationship. This suggests that higher population density is weakly associated with lower migration percentages.

```python

# Scatter plot to visualize the relationship between population density and migration percentage
plt.figure(figsize=(10, 6))
plt.scatter(top10_country['pop_density'], top10_country['migration_perc'], alpha=0.7, color='green')
plt.title('Relationship Between Population Density and Migration Percentage', fontsize=14)
plt.xlabel('Population Density', fontsize=12)
plt.ylabel('Migration Percentage', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()

```

```python
# Calculate the correlation between population and migration percentage
correlation_net_migration_population = top10_country['net_migration'].corr(top10_country['population'])

correlation_net_migration_population
```

The correlation between net migration and populaton is approximately -0.3121, r < 0: Negative correlation (as one variable increases, the other tends to decrease).
A slight tendency exists where countries with higher populations have lower net migration. Countries with large populations might experience less net migration relative to smaller countries. However, this relationship is weak, so it's not a consistent trend.

```python
# Scatter plot to visualize the relationship between net migration and migration percentage
plt.figure(figsize=(10, 6))
plt.scatter(top10_country['population'], top10_country['net_migration'], alpha=0.7, color='blue')
plt.title('Relationship Between Net Migration and Population', fontsize=14)
plt.xlabel('population', fontsize=12)
plt.ylabel('Net Migration', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()

```
