```python
import pandas as pd
from pathlib import Path
```

```python
csv_file = Path("migration_population.csv")
```

```python
data_df = pd.read_csv(csv_file)
data_df
```

```python
data_df.columns.tolist()
```

```python
pd.set_option("display.max_columns", None)
data_df.info(verbose=True)
```

```python
columns_to_remove = ['iso3c', 'iso2c','adminregion']  # Replace with the names of the columns you want to remove
data_df = data_df.drop(columns=columns_to_remove, errors='ignore')
data_df
```

```python
data_df=data_df.dropna()
data_df
```

```python
top10_country=data_df.loc[
    data_df['country'].isin([
        'India', 'Mexico', 'Russian Federation', 'China', 'Bangladesh', 'Pakistan', 'Ukraine', 'Philippines', 'Poland', 'United Kingdom'
    ])]
top10_country
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data_df.head()
```

```python
#vis = sns.load_dataset(data_df["year"])
plt.figure(figsize=(10,4))

sns.boxplot(top10_country, x="country", y="net_migration", hue="country")
plt.xticks(rotation=75)
plt.show()
```

```python
#vis = sns.load_dataset(data_df["year"])
plt.figure(figsize=(10,4))

sns.boxplot(top10_country, x="incomeLevel", y="net_migration")
plt.xticks(rotation=75)
plt.show()
```

```python
plt.figure(figsize=(10,4))
sns.stripplot(top10_country, x="incomeLevel", y="net_migration", hue="country")
plt.legend(loc="lower right", ncol=2) #bbox_to_anchor=(.5,1)
plt.show()
```

```python
from scipy.stats import pearsonr

data_df.loc[data_df["incomeLevel"]=="Low income", 'Income_score'] = 1
data_df.loc[data_df["incomeLevel"]=="Lower middle income", 'Income_score'] = 2
data_df.loc[data_df["incomeLevel"]=="Upper middle income", 'Income_score'] = 3
data_df.loc[data_df["incomeLevel"]=="High income", 'Income_score'] = 4

net_corr, net_pval = pearsonr(data_df['Income_score'], data_df['net_migration'])
print('Net Correl: ', net_corr)
print('Net P_Val: ', net_pval)
```

```python
incCorr = data_df["Income_score"].corr(data_df["pop_density"])
incCorr
```
